```python
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
cases = df.nlargest(10, "survived")
sns.barplot(x="age", y="survived", data = cases , palette="viridis")
plt.title("Top 10 Survived passenger")
plt.show()
```

```python
fig = px.violin(df, x= "age", y= "survived",
    box=True, points="all",
    title="Age vs Survived",
    color="survived")
fig.show()
```
