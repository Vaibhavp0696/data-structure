```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding="latin-1")
df
```

```python

plt.figure(figsize=(25, 9))
sns.barplot(x="Product_Category", y= "State", data = df, palette="Set2")
plt.title("Count of products in each category")
plt.show()
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df, palette="Set2")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data=df, x ="Age Group", hue= "Gender")
 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.barplot(data = df, x="Gender", y="Amount" , estimator=sum,  palette="Set2")
plt.title("Total Amount spent by Gender")
plt.show()
```

```python
plt.figure(figsize=(22,6))
ax = sns.barplot(data = df, x="Occupation", y = "Amount", palette="Set2")

plt.title("Total Amount spent by Product Cateogry")
plt.show()
```

```python
ax = sns.relplot(data= df, x = "Product_Category", y= "Occupation", kind="line",height=5.5, aspect=4, hue='Gender')

plt.title("Product Category vs Occupation")
plt.show()
```
