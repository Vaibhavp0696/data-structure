```python
import pandas as pd
```

```python
df = pd.read_csv("country_wise_latest.csv")
df
```

```python
top_10 = df.nlargest(10, 'Confirmed')
sns.barplot(x="Confirmed", y="Country/Region", data=top_10, palette="viridis")
plt.title("Top 10 Countries with highest number of congitmed")
plt.show()
```

```python
top_countries = df.nlargest(10, 'Deaths / 100 Cases')
plt.figure(figsize=(12,6))
sns.barplot(x='Deaths / 100 Cases', y='Country/Region', data=top_countries, palette='Reds')
plt.title('Top 10 Countries by Death Rate')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Confirmed last week', y='1 week change', data=df, marker='o')
plt.title('Weekly Case Growth')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='Confirmed', y='Active', data=df, hue='WHO Region', palette='viridis')
plt.title('Active Case vs Confirmed case')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Deaths / 100 Cases'], bins = 30, kde=True, color='red')
plt.title("Distribution of Death rate")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='WHO Region', y='New cases', data=df, palette='pastel')
plt.xticks(rotation=45)
plt.title('New Cases by WHO Region')
plt.show()
```

```python
plt.figure(figsize=(8,8))
df.groupby('WHO Region')['Confirmed'].sum().plot.pie(autopct='%1.1f%%', cmap='coolwarm', startangle=140)
plt.ylabel('')
plt.title('Confirmed Cases Distribution by WHO Region')
plt.show()
```

```python
sns.pairplot(df[['Confirmed', 'Deaths', 'Recovered', 'Active']], diag_kind='kde', plot_kws={'alpha':0.5})
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.relplot(data=df, x="Deaths", y="Active",hue="WHO Region",height=6)
plt.title("CASES")
plt.show()
```
