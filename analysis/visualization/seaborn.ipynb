```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df
```

```python
sns.color_palette()
```

```python
itanic = sns.load_dataset("titanic")
sns.catplot(data=titanic, x="sex", y="survived", hue="class", kind="bar")
```

survival anf non-survival count by gender

```python
sns.countplot(df, x='alive', hue = 'sex')
plt.title("Survival and non-survival count by")
plt.show()
```

```python
sns.countplot(df, y = 'class', palette="viridis")
plt.title('Number of passenger by class')
plt.show()
```

```python
sns.barplot(df ,x='class', y="survived",palette='coolwarm')
plt.title('survival Rate by class')
plt.show()
```

```python
sns.histplot(df, x='age', bins=30,kde=True , color = 'red')
plt.title("Age Distribution")
plt.show()
```

```python
sns.violinplot(df, x='class', y='age', hue='sex', palette='coolwarm')
plt.title('age distribution by class')
plt.show()
```
